```python
import numpy as np
from tensorflow import keras
```

```python
X = [1]
for i in range(2, 101):
    X.append(X[i-2]+i**2)

y = X[1:]
X.pop()

X = np.array(X, dtype=int).reshape(99, 1, 1)
y = np.array(y, dtype=int).reshape(99, 1, 1)
X, y
```

```python
model = keras.models.Sequential([
    keras.layers.LSTM(1, input_shape=[1, 1])
])
```

```python
opt = keras.optimizers.Nadam()
model.compile(loss="mae",
              optimizer=opt,
              metrics=["accuracy"])
```

```python
history = model.fit(X, y, epochs=100)
```
