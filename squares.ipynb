```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
```

```python
xs = np.arange(1, 101)

y = [1]
for i in range(2, 101):
    y.append(y[i-2]+i**2)

ys = np.array(y, dtype=int)
```

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(xs, y)
```

```python
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(xs.reshape(-1, 1))
xs[0], X_poly[0]
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_poly, ys)
```

```python
pred_y = lin_reg.predict(X_poly)
plt.plot(xs, ys)
plt.plot(xs, pred_y)
```

```python
print(xs[10], ys[10], pred_y[10])
```

```python
test_X = np.arange(101, 151)

y = [101]
for i in range(102, 151):
    y.append(y[i-102]+i**2)
test_y = np.array(y, dtype=int)

poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(test_X.reshape(-1, 1))
```

```python
pred_y = lin_reg.predict(X_poly)
plt.plot(test_X, test_y)
plt.plot(test_X, pred_y)
```
